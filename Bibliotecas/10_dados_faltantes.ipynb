Preencher dados faltantes com alguma outra informação

Usar o dropna talvez nao seja a melhor solução, pois ele pode impactar profundamente no dataset e, com isso gerar resultados não esperados. 

```python
# Utilizando isnull, verifica em cada variavel se o dado e nulo, retornando true ou false.

df_isnull = df.isnull()
df_isnull
```

```python
# Podemos utilizar a função sum() para verificar o total de dados nulos em cada coluna do dataset.

df_isnull_soma = df_isnull.sum()
df_isnull_soma
```

```python
# Podemos também verificar o percentil de dados faltantes em cada coluna do dataset, atraves de uma formula.

df_faltantes_percentil = (df_isnull_soma / len(df['ID'])) * 100
df_faltantes_percentil
```

```python
# Com essas informações podemos verificar qual coluna está mais impactando no dataset e assim tomar uma ação.
# Vamos utilizar a função fillna(), do qual escolhemos qual coluna do dataset queremos substituir e por qual valor.

# Criar um novo dataset espelho do dataset original, para nao perder as informaçoes
df_new = df

# Substituir os dados faltantes da coluna Medal por none, pessoas que não ganharam medalha
df_new['Medal'].fillna('None', inplace = True)

# Substituir os dados faltantes da coluna Age, Height e Weight pela média de idade do dataset
df_new['Age'].fillna(df_new['Age'].mean(), inplace = True)
df_new['Height'].fillna(df_new['Height'].mean(), inplace = True)
df_new['Weight'].fillna(df_new['Weight'].mean(), inplace = True)

df_new.head(100)
```

```python
# Verificamos agora a quantidade em percentil do total de dados NaN.
# Nota: o dataset manteve o mesmo tamanho do original

print((df_new.isnull().sum() / len(df['ID'])) * 100)
```
